```python
def get_triplet_count(arr, d):
    n = len(arr)
    count = 0

    for i in range(n - 2):
        for j in range(i + 1, n - 1):
            for k in range(j + 1, n):
                triplet_sum = arr[i] + arr[j] + arr[k]
                if triplet_sum % d == 0:
                    count += 1

    return count

# Sample Input
arr = [3,3,4,7,8]
d = 5

# Sample Output
result = get_triplet_count(arr, d)
print(result)
```

    category,

    total_stock
FROM (
    SELECT
        p.category,
        p.title,
        SUM(w.quantity) AS total_stock,
        ROW_NUMBER() OVER (PARTITION BY p.category, p.title ORDER BY SUM(w.quantity) DESC) AS rnk
    FROM
        products p
    JOIN
        warehouse w ON p.product_id = w.product_id
    GROUP BY
        p.category, p.title
    HAVING
        SUM(w.quantity) > 10
) AS ranked
WHERE rnk = 1
ORDER BY
    category ASC,
    title ASC,
    total_stock DESC;
