```python

l=[10,23,30,43,5,6,7]
res={n:n**3 if n%2==0 else n//2 for n in l}
print(res)

x=dict([[i,i**3] if i%2==0 else [i//2,i] for i in l]) # type: ignore
print(x)

r={[n//2,n][n%2==0]:n**3 if n%2==0 else n for n in l}
print(r)
```
