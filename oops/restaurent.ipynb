```python
class Ichiraku:
    menu = {
        "veg": {
            1: "FARMENTED RICE(Saddanam)  50",
            2: "Jonna Rotte               40",
            3: "Ragi Mudda                45",
            4: "Chapathi                  40",
            5: "Vegetable Kurma           50",
            6: "Aaloo 65                  45",
            7: "Gongura Chutney           70"
        },
        "non_veg": {
            1: "Chicken Biryani           120",
            2: "Mutton Paya               250",
            3: "Bheja Fry                 200",
            4: "Chapala Pulsu             100",
            5: "Gongura Chicken            90",
            6: "Mutton Biryani            280"
        },
        "noodles": {
            1: "Veg Noodles               120",
            2: "Chicken Chow Noodles      250",
            3: "Ichiraku Ramen (SPECIAL)  200",
            4: "Egg Noodles               100"
        }
    }

    def __init__(self):
        self.cart = {}

    def display_menu(self):
        for category, items in Ichiraku.menu.items():
            print(f"\n{category} Menu:")
            for num, item in items.items():
                print(f"{num}: {item}")

    def add_to_cart(self, category, item_num):
        if category in Ichiraku.menu and item_num in Ichiraku.menu[category]:
            item = Ichiraku.menu[category][item_num]
            price = int(item.split()[-1])  # Extracting price from the menu
            full_item_name = f"{category}_{item_num}"
            
            if full_item_name in self.cart:
                self.cart[full_item_name]["quantity"] += 1
            else:
                self.cart[full_item_name] = {"item": item, "quantity": 1, "price": price}
            print(f"{item} has been added to your cart.")
        else:
            print("This item is not in the menu.")

    def view_cart(self):
        print("\nYour Cart:")
        total_price = 0
        for item, details in self.cart.items():
            print(f"{details['item']} - {details['quantity']} x {details['price']} = {details['quantity'] * details['price']}")
            total_price += details['quantity'] * details['price']
        print(f"\nTotal Price: {total_price}")


# Example usage
restaurant = Ichiraku()

# Display the menu
restaurant.display_menu()

# Add items to cart
veg=int(input("select veg items using numbers 1 to 7"))
restaurant.add_to_cart("veg", veg)
Nveg=int(input("select Non_veg items using numbers 1 to 7"))
restaurant.add_to_cart("non_veg", Nveg)

# View the cart
restaurant.view_cart()
```
